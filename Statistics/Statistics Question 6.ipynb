Q-6. A pharmaceutical company is developing a drug that is supposed to reduce blood
pressure. They conduct a clinical trial with 100 patients and record their blood
pressure before and after taking the drug. The company wants to know if the change
in blood pressure follows a normal distribution.

To know if the change in blood pressure follows a normal distribution or not we have to plot Q-Q plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv('data.csv')
```

```python
df.columns
```

```python
BP_Before=df[' Blood Pressure Before (mmHg)']
```

```python
BP_After=df[' Blood Pressure After (mmHg)']
```

```python
diff=BP_Before-BP_After
```

```python
plt.hist(diff)
plt.xlabel('Change in Blood Pressure')
plt.ylabel('Frequency')
plt.title('Histogram of Blood Pressure Changes')
plt.show()
```

```python
stats.probplot(diff, dist="norm", plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

#### Observation:
we see the upper end of Q-Q plot is slightly or negatively deviating from the straight line thus making it a perfect fit for Normal Distribution.

Thus, the change in blood pressure follows a normal distribution.
