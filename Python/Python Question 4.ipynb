Write a program to download the data from the link given below and then read the data and convert the into the proper structure and return it as a CSV file.

```python
import pandas as pd
import requests
response = requests.get('https://data.nasa.gov/resource/y77d-th95.json')
data=response.json()
df = pd.DataFrame(data) #converted the json data in pandas dataframe
df.to_csv("output.csv",index=False) #saved the data in csv format
```
