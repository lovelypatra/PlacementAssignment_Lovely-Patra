Insights to be drawn -

● Get all the earth meteorites co-ordinates who fell before the year 1970
● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('nasa.csv')
```

```python

pd.set_option("display.max_rows",None)
```

```python
df.head()
```

```python
df.info()
```

```python
df['fall'].unique()
```

```python
meteorites_fell_before_2000=df[(df['year'] < 2000) & (df['fall'] == 'Fell')]
```

```python
mereorites_fell_before_1970=df[(df['year'] < 1970) & (df['fall'] == 'Fell')]
```

```python
meteorites_mass_more_than_10000kg= df.loc[df['mass'] > 10000]
```

```python
plt.figure(figsize=(15, 20),dpi=200)
plt.suptitle('Distribution of Earth Meteorites by Year (Before 2000)',fontsize = 20, fontweight = 'bold',alpha = 1, y = 1.01)
plt.hist(meteorites_fell_before_2000['year'].astype(int), bins=30)
plt.xlabel('Year', fontdict={'fontsize':12,'fontweight':'bold'})
plt.ylabel('Count', fontdict={'fontsize':12,'fontweight':'bold'})
plt.tight_layout()
plt.show()
```

```python

plt.suptitle('Earth Meteorites Coordinates (Before 1970)',fontsize = 20, fontweight = 'bold',alpha = 1, y = 1.01)
plt.scatter(mereorites_fell_before_1970['reclong'], mereorites_fell_before_1970['reclat'], color='yellow', alpha=0.6)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 20),dpi=200)
plt.suptitle('Distribution of Earth Meteorites Mass (Mass > 10000 kg)',fontsize = 20, fontweight = 'bold',alpha = 1, y = 1.01)
plt.hist(meteorites_mass_more_than_10000kg['mass'].astype(float) / 1000, bins=20, color='red')
plt.xlabel('Mass (kg)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
